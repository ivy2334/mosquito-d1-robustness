```python
#project directory
import os

BASE_PATH = "/content/drive/MyDrive/mosquito-robustness"

os.makedirs(BASE_PATH, exist_ok=True)

print("Project directory ready:", BASE_PATH)
```

```python
#project directory created
 folders = [
    "data",
    "models",
    "results",
    "configs"
]

for f in folders:
    os.makedirs(os.path.join(BASE_PATH, f), exist_ok=True)

print("Folders created.")
```

```python
#config file
config_text = """
data:
  image_size: 192
  train_split: 0.8
  test_split: 0.2

normalization:
  mean: [0.485, 0.456, 0.406]
  std:  [0.229, 0.224, 0.225]

training:
  epochs: 20
  batch_size: 32
  optimizer: adam
  learning_rate: 1e-4

model:
  freeze_backbone: true
  train_last_layer_only: true

noise:
  sigmas: [0.0, 0.05, 0.1, 0.15, 0.2]
"""

config_path = os.path.join(BASE_PATH, "configs/config.yaml")

with open(config_path, "w") as f:
    f.write(config_text)

print("Config file saved at:", config_path)
```

```python
#verification
import yaml

with open(config_path, "r") as f:
    config = yaml.safe_load(f)

print(config)
```

```python
#seed to be used throughout the project
import random
import numpy as np
import torch

def set_seed(seed=42):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)

    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False

set_seed(42)

print("Seed fixed with deterministic behavior.")
```
